### Q1 :- Print the given strings as per stated format.

**Given strings**:
```
"Data" "Science" "Mentorship" "Program"
"By" "CampusX"
```
**Output**:
```
Data-Science-Mentorship-Program-started-By-CampusX
```

Concept- [Seperator and End]

````python
print("Data" , "Science" , "Mentorship" , "Program" , sep="-" , end="-started-")
print("By" , "CampusX" , sep="-")
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
ax = int(input("Enter x-coordinate of 1st point"))
ay = int(input("Enter y-coordinate of first point"))
bx = int(input("Enter x-coordinate of 2nd point"))
by = int(input("Enter y-coordinate of 2nd point"))

print(((ax-bx)**2+(ay-by)**2)**(1/2))
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
heads = int(input("Enter number of heads"))
legs = int(input("Enter number of legs"))

for i in range(heads,-1,-1):
    if ((i*4)+((heads-i)*2)) == legs:
        print("number of dogs are" , i , "and number of chickens are" , heads-i)
    else:
        continue
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
num = int(input("Enter the number"))
sum = 0

for i in range(1,num+1):
    sum += i**2

print(sum)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
a = int(input("Enter 1st number"))
b = int(input("Enter 2nd number"))
n = int(input("Enter n"))

d = b-a

print("An = " , a+(n-1)*d)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
a = int(input("Enter 1st number's numerator"))
b = int(input("Enter 1st number's denominator"))
c = int(input("Enter 2nd number's numerator"))
d = int(input("Enter 2nd number's denominator"))

sum = (a*d+c*b)/(b*d)

print(sum)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
ht = int(input("Enter height of tank"))
lt = int(input("Enter length of tank"))
bt = int(input("Enter breadth of tank"))

hg = int(input("Enter height of glass"))
rg = int(input("Enter radius of glass"))

vt = ht*lt*bt
vg = (22/7)*(rg**2)*(hg)

print(vt//vg)
````
